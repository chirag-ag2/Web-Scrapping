### Scrap the Data From Web

```python
import bs4
import requests
names=[]
ratings=[]
headings=[]
date=[]
config=[]
verification=[]
review=[]
base_url="https://www.amazon.in/product-reviews/B0791YHVMK/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&filterByStar=all_stars&reviewerType=all_reviews&sortBy=recent&pageNumber=pagenumbers"
for j in range(500):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
    }
    url=base_url.replace("pagenumbers",f'{j+1}')
    print(f"--> Fetching Page {j+1}....")
    r=requests.get(url,headers=headers)
    print(f"--> Status Code {r.status_code}.")
    soup=bs4.BeautifulSoup(r.text)
    x=soup.find_all("div",class_="a-section review aok-relative")
    for i in x:
        names.append(i.find("span",class_="a-profile-name").text.strip())
        ratings.append(i.find("span",class_="a-icon-alt").text.strip())
        headings.append(i.find("a",class_="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold").text.strip())
        date.append(i.find("span",class_="a-size-base a-color-secondary review-date").text.strip())
        config.append(i.find("a",class_="a-size-mini a-link-normal a-color-secondary").text.strip())
        if i.find("span",class_="a-size-mini a-color-state a-text-bold"):
            verification.append("Verified")
        else:
            verification.append("Non Verified")
        review.append(i.find("div",class_="a-row a-spacing-small review-data").get_text(separator="\n").strip())
    print(f"--> Fetched Page {j+1}.")
print("--> Website Scraped.")
```

### Create a DataFrame

```python
import pandas as pd
df=pd.DataFrame({
    'Name':names,
    'Date and Location':date,
    'Configuration':config,
    'Verification':verification,
    'Rating':ratings,
    'Heading':headings,
    'Review':review
},index=range(1,5001))
```

### Saving it On Your Machine

```python
df.to_csv("amazon_scrapping_karan.csv")
df.to_html("amazon_scrapping_karan.html")
```

### A little Wrangling....

```python
df["Rating"]=df["Rating"].apply(lambda x:float(x[:3]))
```

```python
df["Date and Location"]=df["Date and Location"].apply(lambda x:x[21:])
```

```python
df["Date and Location"]=df["Date and Location"].astype("datetime64")
```

```python
df["Verification"]=df["Verification"]=="Verified"
```

```python
df["Configuration"]=df["Configuration"].apply(lambda x:x[15:])
```

```python
df[["Heading","Review"]]=df[["Heading","Review"]].astype("str")
```

```python
df.rename(mapper={'Date and Location':'Date',"Verification":'Verified'},axis=1,inplace=True)
```
