```python
import requests
import os
from bs4 import BeautifulSoup
import re
req=requests.get('https://www.sanfoundry.com/')
soup=BeautifulSoup(req.text)
all_links=soup.find('div',class_='entry-content').find_all('a')
d={}
for i in all_links:
    d[i.text]=i.get('href')
# 452 keys
# 1 to 10, 11 to 20,21 to 30
start=int(input('From Key:'))
stop=int(input('To Key:'))
sid=int(input('Folder Number:'))
print(f'Scrapping from {start} to {stop}.')
stop+=1
for subject in list(d.keys())[start:stop]:
    if re.match('1000.*?MCQs',subject):
        name_sub="_".join(subject.split()[1:-1])
        link_sub=d[subject]
        r_sub=requests.get(link_sub)
        soup_sub=BeautifulSoup(r_sub.text)
        sub=soup_sub.find('div',class_='entry-content')
        print(f'--> Writing -->',subject)
        for topic,table in zip(sub.find_all('h4'),sub.find_all('table')):
            print("\t--> Writing -->",topic.text)
            t_name='-'.join(re.split(';|,|\*|\n|/|\\|\.|:','_'.join(topic.text.split())))
            try:
                os.makedirs(f"SF\\{sid}. {name_sub}\\{t_name}\\Questions")
                os.makedirs(f"SF\\{sid}. {name_sub}\\{t_name}\\Answers")
            except OSError:
                pass
            for st in table.find_all('a'):
                questions="----Project By Karan Agrawal & Chirag Agarwal----\n\n\n"
                answers="----Project By Karan Agrawal & Chirag Agarwal----\n\n\n"
                sub_topic=st.text
                print("\t\t--> Writing -->",sub_topic)
                req_st=requests.get(st.get('href'))
                soup_st=BeautifulSoup(req_st.text)
                for que in soup_st.find('div',class_='entry-content').find_all('p')[1:-3]:
                    questions+="\n"+que.text.split('View Answer')[0]+"\n"
                for ans in soup_st.find('div',class_='entry-content').find_all('div',class_='collapseomatic_content'):
                    answers+="\n"+ans.text+"\n"
                f_name='-'.join(re.split(';|,|\*|\n|/|\\|\.|:',('_'.join(sub_topic.split()))))
                try:
                    with open(f"SF\\{sid}. {name_sub}\\{t_name}\\Questions\\{f_name}.txt",'w',encoding='utf-8') as fp:
                        questions+="\n\n----Project By Karan Agrawal and Chirag Agarwal----\n"
                        fp.write(questions)
                    with open(f"SF\\{sid}. {name_sub}\\{t_name}\\Answers\\{f_name}.txt",'w',encoding='utf-8') as fp:
                        answers+="\n\n----Project By Karan Agrawal and Chirag Agarwal----\n"
                        fp.write(answers)
                except FileNotFoundError:
                    print('File Not Found Error Handled')
        sid+=1
```
